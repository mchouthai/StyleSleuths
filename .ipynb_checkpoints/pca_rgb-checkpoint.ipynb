```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.stats import stats
import matplotlib.image as mpimg
```

```python
images_rgb = np.load("data_resized/images_rgb.npy")
```

```python
images_rgb.shape
```

```python
images_rgb
```

```python
#splitting up the data into blues, greens and reds. 
blues = images_rgb[:,:,:,2]
greens = images_rgb[:,:,:,1]
reds = images_rgb[:,:,:,0]
```

```python
blues = blues/63
greens = greens/63
reds = reds/63
```

```python
length = len(images_rgb)
```

```python
#conducting pca on the blue, green, red data separately
pcaB = []
pcaG = []
pcaR = []

transformed_blue = []
transformed_green = []
transformed_red = []

for i in range(length):
    pcaB.append(PCA(n_components=30))
    pcaG.append(PCA(n_components=30))
    pcaR.append(PCA(n_components=30))


for i in range(length):
    pcaB[i].fit(blues[i])
    transformed_blue.append(pcaB[i].transform(blues[i]))
    
    pcaG[i].fit(greens[i])
    transformed_green.append(pcaG[i].transform(greens[i]))

    pcaR[i].fit(reds[i])
    transformed_red.append(pcaR[i].transform(reds[i]))


   
```

```python
print(sum(pcaB[0].explained_variance_ratio_))
```

```python
#reconstructing the images

reconBlues = []
reconGreens = []
reconReds = []

for i in range(length):
    reconBlues.append(pcaB[i].inverse_transform(transformed_blue[i]))
    reconGreens.append(pcaG[i].inverse_transform(transformed_green[i]))
    reconReds.append(pcaR[i].inverse_transform(transformed_red[i]))
```

```python
import cv2
```

```python
reduced_rgb = []

for i in range(length):
    reduced_rgb.append(cv2.merge((transformed_red[i], transformed_green[i], transformed_blue[i])))
```

```python
reduced_rgb = np.asarray(reduced_rgb)
```

```python
np.save('reduced_data', reduced_rgb)
```
